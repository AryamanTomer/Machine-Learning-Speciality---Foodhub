```python
#Question 13: The company wants to provide a promotional offer in the advertisement of the restaurants. The condition to get the offer is that the restaurants must have a rating count of mroe than 50 and the average rating shouuld be greater than 4. Find the restaurants fulfilling the criteria to get the promotional offer.
df_rated = df[df['rating'] != 'Not given'].copy()

df_rated['rating'] = pd.to_numeric(df_rated['rating'])
df_rated['count_rating'] = df_rated['rating'].groupby(df['restaurant_name']).transform('count')
df_rated['avg_rating'] = df_rated['rating'].groupby(df['restaurant_name']).transform('mean')
df_rated = df_rated[df_rated['avg_rating'] > 4.00]
df_rated = df_rated[df_rated['count_rating'] > 50]
df_rating_total = df_rated[['restaurant_name']]
df_rating_total.head(10)
```

```python
#Question 14: The company charges the restaurant 25% on the orders having cost greater than 20 $ and 15% on the orders having cost greater than 5 dollars. Find the net revenue generated by the company across all orders.
df_order = df[['order_id','cost_of_the_order']].copy()
df_order['25'] = df_order['cost_of_the_order'].apply(lambda x: x*.25 if x > 20 else 0)
df_order['15'] = df_order['cost_of_the_order'].apply(lambda x: x*.15 if x > 5 and x <= 20 else 0)
df_order['Total'] = df_order['25'] + df_order['15']
df_order['Net'] = df_order['Total'].sum()
df_order = df_order[['Net']]
df_order = df_order.drop_duplicates(subset=['Net'])
print('The net revenue is around')
df_order
```

```python
#Question 15: The company wants to analyze the total time required to deliver the food. What percentage of orders take more than 60 minutes to get delivered from the time the order is places?
df_deliver = df
df_deliver['total_time'] = df['food_preparation_time'] + df['delivery_time']
df_deli = df_deliver[df_deliver['total_time']>60]
print('% of orders that take more than 60 minutes:')
print(df_deli['order_id'].count()/df_deliver['order_id'].count() * 100)
```

```python
#Question 16: The company wants to analyze the delivery time of the orders on the weekdays and the weekends. How does the mean delivery time vary during weekdays and weekends?
print('The mean delivery time on weekdays is around',
     round(df[df['day_of_the_week'] == 'Weekday']['delivery_time'].mean()),
     'minutes')
```
